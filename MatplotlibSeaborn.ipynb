```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sbn
```

```python
%matplotlib inline
```

```python
yasListesi=[10,20,30,40,50,60,70,80,90]
```

```python
kiloListesi=[20,60,70,75,80,74,77,72,70]
```

```python
numpyKiloListesi=np.array(yasListesi)
```

```python
numpyYasListesi=np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
figure2=plt.figure()
eksen1=figure2.add_axes([0.1,0.1,0.9,0.9])
eksen1.plot(numpyKiloListesi,numpyYasListesi,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2=figure2.add_axes([0.2,0.5,0.3,0.3])
eksen2.bar(numpyYasListesi,numpyKiloListesi,color="skyblue")
eksen2.set_xlabel("x ekseni")
eksen2.set_ylabel("y ekseni")
eksen2.set_title("küçük grafik başlık")
#plt.show()
```

```python
figure2.savefig("Figure.png",dpi=150)
```

```python
data=[[21,5000],[25,6000],[30,10000],[35,15000]]
```

```python
yeniData=np.array(data)
```

```python
yeniData
```

```python
dataFrame=pd.DataFrame(data=yeniData,index=["Ahmet","Ali","Mehmet","Veli"],columns=["Yas","Maas"])
```

```python
dataFrame
```

```python
dataFrame.plot(kind="bar", figsize=(8, 5))
plt.xlabel("İsimler")
plt.ylabel("Maaşlar")
plt.title("Yaş ve Maaş Dağılımı")
```

```python
plt.figure(figsize=(6, 5))
sbn.boxplot(data=dataFrame)
plt.title("Yaş ve Maaş Dağılımı (Boxplot)")
plt.ylabel("Değerler")
plt.legend()
```

```python
plt.figure(figsize=(7, 5))
sbn.kdeplot(dataFrame["Yas"], fill=True, color="blue", label="Yaş")
sbn.kdeplot(dataFrame["Maas"], fill=True, color="red", label="Maaş")

plt.title("Yaş ve Maaş Dağılım Grafiği")
plt.xlabel("Değerler")
plt.ylabel("Yoğunluk")
plt.legend()
```

```python
plt.figure(figsize=(7, 5))
sbn.histplot(dataFrame["Yas"], kde=True, color="blue", label="Yaş", bins=10, alpha=0.6)
sbn.histplot(dataFrame["Maas"], kde=True, color="red", label="Maaş", bins=10, alpha=0.6)

plt.title("Yaş ve Maaş Dağılım Grafiği (Distplot)")
plt.xlabel("Değerler")
plt.ylabel("Frekans")
plt.legend()
```

```python
plt.figure(figsize=(6, 4))
sbn.heatmap(dataFrame, annot=True, cmap="viridis", linewidths=0.5, fmt=".1f")

plt.title("Yaş ve Maaş Isı Haritası (Heatmap)")
plt.xlabel("Özellikler")
plt.ylabel("İsimler")
plt.show()
```
